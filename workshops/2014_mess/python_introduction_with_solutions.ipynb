In this exercise we are going to simulate a dice and roll it as often as it takes to get a 6. The first step is import a function from the random module. This is explained in more detail later on.

1. Write a short program that throws the dice until a 6 has been thrown and count how many tries it took.
2. **[advanced]** Throw until two successive (not parallel!) throws yield 6. Count the amount of throws it took.
3. **[expert]** Count the amount of tries it takes when two dice are thrown in parallel to throw each possible combination [(1, 1), (1, 2), ..] at least once.
4. **[expert]** Throw until *N* successive (not parallel!) throws yield 6. Count the amount of throws it took.

**There is a STOP button in the toolbar in case you accidentially created an infinite loop!**


Documentation: [while-statement](https://docs.python.org/3/tutorial/introduction.html#first-steps-towards-programming)


````python
from random import randint

# Default Exercise.
tries = 0

while True:
    tries += 1
    if randint(1, 6) == 6:
        break

print "It took %i tries!" % tries


# Advanced Exercise.
tries = 0
previous_throw = None

while True:
    tries += 1
    this_throw = randint(1, 6)
    if this_throw == 6 and previous_throw == 6:
        break
    previous_throw = this_throw

print "It took %i tries to get two successive 6ths!" % tries


# First expert exercise. This solution illustrates some features of the standard library.
# Could also be solved with a set which is likely easier.
import collections
import itertools

counter = collections.Counter()

for permutation in itertools.permutations(range(1, 7), 2):
    counter[permutation] += 0

tries = 0
while True:
    tries += 1
    counter[(randint(1, 6), randint(1, 6))] += 1
    if all(counter.values()):
        break

print "It took %i tries to get all combinations!" % tries


# First expert exercise. This time with a set.
tries = 0
unique_combinations = set()
while True:
    tries += 1
    unique_combinations.add((randint(1, 6), randint(1, 6)))
    if len(unique_combinations) >= 36:
        break
print "It took %i tries to get all combinations!" % tries


# Second export exercise.
tries = 0
N = 4
last_throws = collections.deque([0] * N, maxlen=N)
while True:
    tries += 1
    last_throws.append(randint(1, 6))
    if set(last_throws) == {6}:
        break
print "It took %i tries to get %i successive 6ths!" % (tries, N)
````

### Exercises 10.3

* **[advanced]** Write a resursive function returning the first n Fibonacci numbers.
* **[expert]** The version above breaks down for large `n` due to Python's recursion limit. Write a version utilizing a generator that does not have this behaviour.

````python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    
print "First 10 Fibonacci numbers:", [fib(i) for i in range(10)]


## Expert Exercise
def fib(n):
    count = 0
    a, b = 0, 1
    while count < n:
        count += 1
        yield a
        a, b = b, a + b
        
print "First 10 Fibonacci numbers:", list(fib(10))
````

## Exercise 12: Basemap

Basemap is an extension of matplotlib for plotting maps.

Write a function *mymap* that:

* Arguments: latitude and longitude
* Plots a red circle on that location.
* Use the *moll* projection, it just requires *lon\_0* (central longitude of projection) as an additional argument.

Documentation: [Mollweide](http://matplotlib.org/basemap/users/moll.html)

````python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

def mymap(latitude, longitude):
    m=Basemap(projection="moll", lon_0=longitude)
    m.drawcoastlines()
    m.drawmapboundary()
    m.fillcontinents()
    lon, lat = longitude, latitude
    x,y=m(lon, lat)
    m.scatter(x, y, color="red", zorder=2)
    plt.show()

latitude = 10
longitude = 50

mymap(latitude, longitude)
````

````python
from pandas.io.parsers import read_csv
import obspy

events = read_csv("./data/python/iris_all_events_since_2012_01_01.txt",
                  sep="|")

# Convert times to UTCDateTime objects.
events["Time"] = events["Time"].apply(obspy.UTCDateTime)

events[:5]
````

To discover relations between columns use `matplotlib.pyplot.scatter()` to plot values against each other. You can also use the values to recreate the map plot above with more data (basemap does not like pandas data series objects - convert to lists first!).

````python
plt.figure(figsize=(12, 5))
plt.scatter(events["Longitude"], events["Latitude"], alpha=0.2)
plt.show()


plt.figure(figsize=(12, 5))
plt.scatter(events["Longitude"], events["Magnitude"], alpha=0.2)
plt.xlabel("Longitude")
plt.ylabel("Magnitude")
plt.show()


# Map plot once again
from mpl_toolkits.basemap import Basemap

plt.figure(figsize=(12, 5))

# Convert depth to colors.
max_depth = events["Depth"].max()
colors = events["Depth"].apply(lambda x: plt.cm.RdYlGn_r(x / max_depth))
        
m=Basemap(projection="robin", lon_0=0)
m.drawcoastlines()
m.drawmapboundary()
m.fillcontinents()
x, y = m(list(events["Longitude"]), list(events["Latitude"]))
m.scatter(x, y, color=colors, zorder=2, alpha=0.2)
plt.title("All IRIS events since 2012/1/1")
plt.show()
````
